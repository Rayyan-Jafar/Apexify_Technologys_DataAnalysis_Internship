```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("spscientist/students-performance-in-exams")

print("Path to dataset files:", path)
```

```python
import pandas as pd
df = pd.read_csv(path + '/StudentsPerformance.csv')
```

```python
df.head()
```

```python
df["Average"] = df[["math score", "reading score", "writing score"]].mean(axis=1)
```

```python

top_student = df.loc[df["Average"].idxmax()]
```

```python
print("All Students with their Average Scores:\n", df[["math score", "reading score", "writing score", "Average"]].head())

print("\nStudent with the Highest Average:\n", top_student)
```

```python
import matplotlib.pyplot as plt

# Bar chart of average scores (first 20 students)
plt.figure(figsize=(12,6))
plt.bar(df.index[:20], df["Average"][:20], color="skyblue")
plt.title("Average Scores of First 20 Students")
plt.xlabel("Student Index")
plt.ylabel("Average Score")
plt.show()
```

```python
plt.figure(figsize=(8,6))
df[["math score", "reading score", "writing score"]].boxplot()
plt.title("Score Distribution by Subject")
plt.ylabel("Score")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(df["Average"], bins=15, color="lightgreen", edgecolor="black")
plt.title("Distribution of Average Scores")
plt.xlabel("Average Score")
plt.ylabel("Number of Students")
plt.show()
```
